```python
# stores years in array of name cols.
cols=['2001', '2002', '2003', '2004', '2005', '2006', '2007',
       '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016',
       '2017', '2018', '2019', '2020', '2021', '2022']
```

# Plotting Histogram

```python
for i in cols:
        fig = plt.figure(figsize=(5,5))
        plt.hist(df[i],color='g',bins=10)
        plt.ylabel("Country")
        plt.xlabel("Data Values")
        plt.title(i)
        plt.show()
```

# plotting bar chart

```python
years=cols
total_values=df[years].sum()
```

```python
plt.figure(figsize=(25,25))
plt.bar(years,total_values)
plt.ylabel('Total Values', size=10)
plt.xlabel('Year', size=20)
plt.title('Total Values per Year', size=20)
plt.show()
```

# Compare data of all year with base year 2001

```python
top_countries=df.sort_values(by="2001").head(10)
top_countries
```

```python
country_by_2001=top_countries.set_index('Country Name').T
country_by_2001
```

```python
for country_name, data_values in country_by_2001.iterrows():
    fig = plt.figure(figsize=(10, 5))
    sb.barplot(x=data_values.index, y=data_values.values)
    plt.xlabel('Countries')
    plt.ylabel('Data Values')
    plt.title(f"{country_name} - Data Values from 2001 to 2022")
    plt.xticks(rotation=90)
    plt.show()
```

# Compare data of all year with base year 2022

```python
country_by_2022 = df.sort_values(by='2022').head(10)
country_by_2022
```

```python
country_by_2022_t = country_by_2022.set_index('Country Name').T
for country_name, data_values in country_by_2022_t.iterrows():
    fig = plt.figure(figsize=(10, 5))
    sb.barplot(x=data_values.index, y=data_values.values)
    plt.xlabel('Country')
    plt.ylabel('Data Value')
    plt.title(f"{country_name} - Data Values from 2001 to 2022")
    plt.xticks(rotation=90)
    plt.show()
```
